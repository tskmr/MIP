```r
library(tidyverse)
t = seq(0, 3, len=100)
cdf = 1-exp(-t^1.7)
d_cdf = data.frame(func = "cdf", t, y = cdf)
d_cdf %>% ggplot() + geom_point(aes(t, y))
```

```r
pdf = 1.7 * t^0.7 * exp(-t^1.7)
d_pdf = data.frame(func = "pdf", t, y = pdf)
d_pdf %>% ggplot() + geom_point(aes(t, y))
```

```r
sf = 1 - cdf
d_sf = data.frame(func = "sf", t, y = sf)
d_sf %>% ggplot() + geom_point(aes(t, y))
```

```r
hf = pdf/ sf
d_hf = data.frame(func = "hf", t, y = hf)
d_hf %>% ggplot() + geom_point(aes(t, y))
```

```r
d = bind_rows(d_cdf, d_pdf, d_sf, d_hf)
d %>% ggplot() + geom_point(aes(t, y))
```
